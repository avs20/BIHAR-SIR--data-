```python
def clean_voter_name_and_father(row):
    if pd.isna(row['voter_name']) or '<br>' not in str(row['voter_name']): return row
    parts = row['voter_name'].split('<br>')
    if len(parts) >= 2:
        row['voter_name'] = parts[0].strip()
        father_part = parts[1].strip()
        for prefix in ['पति का नाम', 'पिता का नाम', 'अन्य']:
            if prefix in father_part:
                father_name = father_part.split(prefix)[-1].lstrip(':').strip()
                if pd.isna(row['father_name']) or row['father_name'] == '': row['father_name'] = father_name
                break
    return row

br_rows = br_rows.apply(clean_voter_name_and_father, axis=1)
br_rows[br_rows['voter_name'].str.contains('<br>', na=False)][['voter_name', 'father_name']].head(10)
```

```python
import re
def clean_house_data(df_subset):
    df_clean = df_subset.copy()
    for idx, row in df_clean.iterrows():
        if idx == 3242:
            df_clean.at[idx, 'father_name'] = 'सुकनी देवी'
            df_clean.at[idx, 'house_number'] = 'पिर्ही'
        elif idx == 212484:
            df_clean.at[idx, 'father_name'] = 'राम कुमार'
            df_clean.at[idx, 'house_number'] = 'वार्ड नं -11'
        elif pd.notna(row['house_number']) and '<br>' in str(row['house_number']):
            parts = row['house_number'].split('<br>')
            if len(parts) >= 2:
                df_clean.at[idx, 'house_number'] = parts[0].strip()
                info_part = parts[1].strip()
                age_match = re.search(r'उम्र\s*:\s*(\d+)', info_part)
                gender_match = re.search(r'लिंग:\s*:\s*([^\s]+)', info_part)
                if age_match: df_clean.at[idx, 'age'] = float(age_match.group(1))
                if gender_match: df_clean.at[idx, 'gender'] = gender_match.group(1)
    return df_clean

br_house_cleaned = clean_house_data(br_house)
br_house_cleaned[['father_name', 'house_number', 'age', 'gender']]
```

```python
def analyze_father_name_br(df):
    br_father = df[df['father_name'].str.contains('<br>', na=False)].copy()
    single_br = br_father[br_father['father_name'].str.count('<br>') == 1]
    multiple_br = br_father[br_father['father_name'].str.count('<br>') > 1]
    
    print(f"Single <br> cases: {len(single_br)}")
    print(f"Multiple <br> cases: {len(multiple_br)}")
    
    if len(multiple_br) > 0:
        print("\nMultiple <br> cases (need manual handling):")
        for idx, row in multiple_br.iterrows():
            print(f"Index {idx}: {row['father_name']}")
    
    return single_br, multiple_br

single_br, multiple_br = analyze_father_name_br(df)
```

```python
def clean_single_br_father(df):
    df_clean = df.copy()
    single_br_mask = df_clean['father_name'].str.contains('<br>', na=False) & (df_clean['father_name'].str.count('<br>') == 1)
    for idx in df_clean[single_br_mask].index:
        parts = df_clean.at[idx, 'father_name'].split('<br>')
        if len(parts) == 2:
            df_clean.at[idx, 'father_name'] = parts[0].strip()
            if pd.isna(df_clean.at[idx, 'house_number']) or df_clean.at[idx, 'house_number'] == '': df_clean.at[idx, 'house_number'] = parts[1].strip()
    return df_clean

df = clean_single_br_father(df)
single_br_after, multiple_br_after = analyze_father_name_br(df)
```

```python
def clean_multiple_br_father(df):
    df_clean = df.copy()
    special_indices = [105917, 113676, 176254]
    
    multiple_br_indices = [35826, 63302, 74883, 88091, 103207, 105917, 113676, 136789, 176254, 202299, 279822]
    
    for idx in multiple_br_indices:
        if idx in df_clean.index:
            full_text = df_clean.at[idx, 'father_name']
            if idx in special_indices:
                parts = full_text.split('<br>')
                if len(parts) >= 3:
                    df_clean.at[idx, 'father_name'] = parts[1].split(':')[-1].strip()
                    house_part = parts[2].split(':')[-1].strip() if ':' in parts[2] else parts[2].strip()
                    if pd.isna(df_clean.at[idx, 'house_number']) or df_clean.at[idx, 'house_number'] == '': df_clean.at[idx, 'house_number'] = house_part
                    if len(parts) >= 4:
                        age_gender_part = parts[3]
                        age_match = re.search(r'उम्र\s*:\s*(\d+)', age_gender_part)
                        gender_match = re.search(r'लिंग:\s*:\s*([^\s]+)', age_gender_part)
                        if age_match and (pd.isna(df_clean.at[idx, 'age']) or df_clean.at[idx, 'age'] == ''): df_clean.at[idx, 'age'] = float(age_match.group(1))
                        if gender_match and (pd.isna(df_clean.at[idx, 'gender']) or df_clean.at[idx, 'gender'] == ''): df_clean.at[idx, 'gender'] = gender_match.group(1)
            else:
                parts = full_text.split('<br>')
                if len(parts) >= 3:
                    df_clean.at[idx, 'father_name'] = parts[0].strip()
                    house_part = parts[1].split(':')[-1].strip() if ':' in parts[1] else parts[1].strip()
                    if pd.isna(df_clean.at[idx, 'house_number']) or df_clean.at[idx, 'house_number'] == '': df_clean.at[idx, 'house_number'] = house_part
                    age_gender_part = parts[2]
                    age_match = re.search(r'उम्र\s*:\s*(\d+)', age_gender_part)
                    gender_match = re.search(r'लिंग:\s*:\s*([^\s]+)', age_gender_part)
                    if age_match and (pd.isna(df_clean.at[idx, 'age']) or df_clean.at[idx, 'age'] == ''): df_clean.at[idx, 'age'] = float(age_match.group(1))
                    if gender_match and (pd.isna(df_clean.at[idx, 'gender']) or df_clean.at[idx, 'gender'] == ''): df_clean.at[idx, 'gender'] = gender_match.group(1)
    
    return df_clean

df = clean_multiple_br_father(df)
single_br_final, multiple_br_final = analyze_father_name_br(df)
```

```python
def standardize_ward_numbers(house_number):
    if pd.isna(house_number): return house_number
    house_str = str(house_number).strip()
    if 'वार्ड' in house_str:
        ward_match = re.search(r'वार्ड[^\d]*(\d+)', house_str)
        if ward_match:
            ward_num = int(ward_match.group(1))
            return f"वार्ड-{ward_num:02d}"
    return house_number

df['house_number'] = df['house_number'].apply(standardize_ward_numbers)
df['house_number'].value_counts().head(20)
```

```python
def standardize_makan_numbers(house_number):
    if pd.isna(house_number): return house_number
    house_str = str(house_number).strip()
    if 'मकान संख्या' in house_str and ':' in house_str:
        parts = house_str.split(':', 1)
        if len(parts) == 2: return parts[1].strip()
    return house_number

df['house_number'] = df['house_number'].apply(standardize_makan_numbers)
df[df['house_number'].str.contains('मकान', na=False)]['house_number'].value_counts().head(10)
```

```python
def standardize_makan_numbers(house_number):
    if pd.isna(house_number): return house_number
    house_str = str(house_number).strip()
    if 'मकान' in house_str:
        makan_match = re.search(r'मकान[^\d]*(\d+)', house_str)
        if makan_match: return makan_match.group(1)
        elif ':' in house_str:
            parts = house_str.split(':', 1)
            if len(parts) == 2: return parts[1].strip()
    return house_number

df['house_number'] = df['house_number'].apply(standardize_makan_numbers)
df[df['house_number'].str.contains('मकान', na=False)]['house_number'].value_counts().head(10)
```
